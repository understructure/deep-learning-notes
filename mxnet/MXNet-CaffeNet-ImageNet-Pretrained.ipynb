```python
def download(url, prefix = ""):
    file_name = prefix+url.split("/")[-1]
    # Download the file from `url` and save it locally under `file_name`:
    with urllib.request.urlopen(url) as response, open(file_name, 'wb') as out_file:
        shutil.copyfileobj(response, out_file)

path="http://data.dmlc.ml/models/imagenet/caffenet/"
download(path+"caffenet-0000.params")
download(path+"caffenet-symbol.json")

# for future reference...
path="http://data.dmlc.ml/models/imagenet/"
download(path+"synset.txt")
```
